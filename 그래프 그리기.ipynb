```python
import re
import numpy as np
import matplotlib.pyplot as plt

# 로그 파일 읽기
with open('loss_256+.txt', 'r',encoding='utf-8') as f:
    log_text = f.read()

# Mean loss와 Valid_Mean_loss 추출
mean_loss_values = re.findall(r"Mean loss: ([\d.]+)", log_text)
valid_mean_loss_values = re.findall(r"Valid_Mean_loss: ([\d.]+)", log_text)

# 문자열을 숫자로 변환
mean_loss_values = [float(val) for val in mean_loss_values]
valid_mean_loss_values = [float(val) for val in valid_mean_loss_values]

# 그래프 그리기
plt.figure(figsize=(16,9))

plt.plot(mean_loss_values, label='train_loss', marker='o', color='blue')
plt.plot(valid_mean_loss_values, label='valid_loss', marker='o', color='red')

plt.title('train_loss and valid_loss of 256+')
plt.xlabel('EPOCH')
plt.ylabel('LOSS')
plt.legend()
plt.grid(True)
plt.show()
```
